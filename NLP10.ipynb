```python
!pip -q install datasets seqeval matplotlib scikit-learn

from datasets import load_dataset

# Use the modern Hugging Face mirror for CoNLL2003
dataset = load_dataset("Babelscape/wikineural", "default")

# This dataset is a multi-lingual NER one; we'll just keep English
dataset = dataset["train_en"].train_test_split(test_size=0.2, seed=42)
dataset["validation"] = dataset["test"]

print(dataset["train"][0])
```

```python
from collections import Counter, defaultdict
import math

# Extract sentences (tokens + tags)
def to_sents(ds, tag_names):
    return [
        (ex["tokens"], [tag_names[t] for t in ex["ner_tags"]])
        for ex in ds
    ]

# Manually define tag_names as the ClassLabel metadata seems to be missing
tag_names = ['O', 'B-PER', 'I-PER', 'B-ORG', 'I-ORG', 'B-LOC', 'I-LOC', 'B-MISC', 'I-MISC']

train_sents = to_sents(dataset["train"], tag_names)
valid_sents = to_sents(dataset["validation"], tag_names)
test_sents  = to_sents(dataset["test"], tag_names)

# Build word frequency
freq = Counter(w for sent, _ in train_sents for w in sent)

UNK = "<UNK>"
min_freq = 2

def normalize_train_word(w):
    return w if freq[w] >= min_freq else UNK

def normalize_eval_word(w, train_vocab):
    return w if w in train_vocab else UNK

train_vocab = {w for w, c in freq.items() if c >= min_freq}
train_vocab.add(UNK)

all_tags = sorted({t for _, tags in train_sents for t in tags})
tag2id = {t: i for i, t in enumerate(all_tags)}
id2tag = {i: t for t, i in tag2id.items()}

def normalize_dataset(sents, mode="train"):
    norm = []
    for words, tags in sents:
        if mode == "train":
            words = [normalize_train_word(w) for w in words]
        else:
            words = [normalize_eval_word(w, train_vocab) for w in words]
        norm.append((words, tags))
    return norm

train_sents_n = normalize_dataset(train_sents, "train")
valid_sents_n = normalize_dataset(valid_sents, "eval")
test_sents_n  = normalize_dataset(test_sents, "eval")

print("Training sentences:", len(train_sents_n))
print("Unique words:", len(train_vocab))
print("Tags:", all_tags)
```

```python
from collections import defaultdict, Counter
import math

alpha = 1.0  # Laplace smoothing

# Count frequencies
start_counts = Counter()
trans_counts = defaultdict(Counter)
emit_counts  = defaultdict(Counter)
tag_counts   = Counter()

for words, tags in train_sents_n:
    if not words:
        continue
    start_counts[tags[0]] += 1
    tag_counts[tags[0]] += 1
    emit_counts[tags[0]][words[0]] += 1
    for i in range(1, len(words)):
        prev_tag, tag = tags[i-1], tags[i]
        trans_counts[prev_tag][tag] += 1
        tag_counts[tag] += 1
        emit_counts[tag][words[i]] += 1

V = len(train_vocab)
T = len(all_tags)

def log(x):
    return -1e9 if x <= 0 else math.log(x)

# Start log-probabilities
start_logp = {}
total_starts = sum(start_counts.values())
for t in all_tags:
    start_logp[t] = log((start_counts[t] + alpha) / (total_starts + alpha * T))

# Transition log-probabilities
trans_logp = {t_prev: {} for t_prev in all_tags}
for t_prev in all_tags:
    total = sum(trans_counts[t_prev].values())
    for t in all_tags:
        trans_logp[t_prev][t] = log((trans_counts[t_prev][t] + alpha) / (total + alpha * T))

# Emission log-probabilities
emit_logp = {t: {} for t in all_tags}
for t in all_tags:
    total = sum(emit_counts[t].values())
    for w in train_vocab:
        emit_logp[t][w] = log((emit_counts[t][w] + alpha) / (total + alpha * V))

print("✅ HMM parameters computed successfully.")
print(f"Number of tags: {T}, Vocabulary size: {V}")
```

```python
from collections import defaultdict

def viterbi_decode(words):
    n = len(words)
    if n == 0:
        return []

    dp = [defaultdict(lambda: -1e18) for _ in range(n)]  # dynamic programming table
    bp = [defaultdict(lambda: None) for _ in range(n)]   # backpointer table

    # Initialization
    w0 = words[0] if words[0] in train_vocab else UNK
    for t in all_tags:
        dp[0][t] = start_logp[t] + emit_logp[t].get(w0, -1e9)

    # Forward pass
    for i in range(1, n):
        wi = words[i] if words[i] in train_vocab else UNK
        for t in all_tags:
            emit_lp = emit_logp[t].get(wi, -1e9)
            best_val = -1e18
            best_prev = None
            for t_prev in all_tags:
                cand = dp[i - 1][t_prev] + trans_logp[t_prev][t] + emit_lp
                if cand > best_val:
                    best_val = cand
                    best_prev = t_prev
            dp[i][t] = best_val
            bp[i][t] = best_prev

    # Backtrack
    last_tag = max(all_tags, key=lambda t: dp[n - 1][t])
    path = [last_tag]
    for i in range(n - 1, 0, -1):
        path.append(bp[i][path[-1]])
    path.reverse()

    return path

# Quick sanity check
sample_words, _ = test_sents[0]
sample_norm = [w if w in train_vocab else UNK for w in sample_words]
pred_tags = viterbi_decode(sample_norm)
print("Sample output (first 10 tokens):")
for w, p in list(zip(sample_words, pred_tags))[:10]:
    print(f"{w:15} -> {p}")
```

```python
from seqeval.metrics import classification_report, f1_score

def predict_dataset(sents_norm, raw_sents): # Removed id2tag parameter as it's no longer needed here
    preds = []
    for (words_norm, _) in sents_norm: # Iterate through normalized sentences
        pred_tags = viterbi_decode(words_norm) # viterbi_decode returns string tags
        preds.append(pred_tags)
    return preds

# Generate predictions
print("Decoding validation set... (this might take a minute)")
# Removed id2tag argument from predict_dataset calls
valid_preds = predict_dataset(valid_sents_n, valid_sents)

print("Decoding test set...")
# Removed id2tag argument from predict_dataset calls
test_preds = predict_dataset(test_sents_n, test_sents)

# Ground truth tags (already in string format from to_sents)
y_true_valid = [tags for _, tags in valid_sents]
y_true_test  = [tags for _, tags in test_sents]

# Debugging: Print the first few examples of y_true and y_pred
print("\nDebugging - First 2 ground truth validation sentences:")
print(y_true_valid[:2])
print("\nDebugging - First 2 predicted validation sentences:")
print(valid_preds[:2])


# Evaluate
print("\n📊 RESULTS:")
print("Validation F1:", round(f1_score(y_true_valid, valid_preds) * 100, 2), "%")
print("Test F1:", round(f1_score(y_true_test, test_preds) * 100, 2), "%")

print("\nValidation classification report:\n", classification_report(y_true_valid, valid_preds))
print("\nTest classification report:\n", classification_report(y_true_test, test_preds))
```
